```python
def merge_all_files(files_list:list) ->pd.DataFrame:
    final_df = pd.DataFrame() 
    for file in files_list:
        temp_df = pd.read_csv(file)
        final_df = pd.concat([final_df, temp_df])
    
    unwanted_cols = ['max_info_centrality', 'avg_info_centrality', 'max_betweenness_centrality',
                    'avg_betweenness_centrality', 'global_efficiency', 'avg_closeness_centrality']
    final_df.drop(unwanted_cols, axis=1, inplace=True)
    
    return final_df
```

```python
small_graphs_feat_df = merge_all_files(small_graphs_features_files)
```
